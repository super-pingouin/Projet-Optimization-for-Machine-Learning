# Convergence Rate calculator 

```python
def compute_convergence_on_mean(results, threshold=0.70):
    """
    Calculate the epoch at which the mean accuracy (averaged over all seeds) 
    for each optimizer and learning rate first reaches or exceeds a given threshold.
    Adds a new key 'mean_acc_epochs_to_threshold' to each optimizer-learning rate entry
    in the results dictionary, indicating the epoch number of convergence.
    If the threshold is never reached, this value is set to NaN.
    """
    for opt_name in results:
        for lr in results[opt_name]:
            mean_acc = results[opt_name][lr]['mean_acc']
            reached_epoch = None
            
            # Iterate through mean accuracies per epoch to find first crossing of threshold
            for i, acc in enumerate(mean_acc):
                if acc >= threshold:
                    reached_epoch = i
                    break
            
            # If threshold never reached, assign NaN
            if reached_epoch is None:
                reached_epoch = np.nan
            
            results[opt_name][lr]['mean_acc_epochs_to_threshold'] = reached_epoch
```

```python
threshold = 0.65
# Compute the number of epochs to reach the accuracy threshold for each optimizer and learning rate
compute_convergence_on_mean(results, threshold=threshold)

def plot_convergence_heatmap(results, threshold=0.6):
    """
    Plot a heatmap showing the number of epochs required for each optimizer-learning rate
    pair to reach the specified accuracy threshold.
    Cells where the threshold was not reached are shown in a distinct lighter color.
    """
    data = []
    # Collect data into a list of dictionaries for DataFrame construction
    for opt_name in results:
        for lr in results[opt_name]:
            # Get the epochs to threshold, or NaN if not attained
            val = results[opt_name][lr].get("mean_acc_epochs_to_threshold", np.nan)
            data.append({"Optimizer": opt_name, "LR": lr, "Epochs to Threshold": val})

    # Convert the collected data to a DataFrame
    df = pd.DataFrame(data)

    # Pivot the DataFrame to have Optimizers as rows and Learning Rates as columns
    pivot = df.pivot(index="Optimizer", columns="LR", values="Epochs to Threshold")
    
    # Create a mask for NaN values (threshold not reached cases)
    mask = pivot.isna()

    # Create a reversed blue colormap for the heatmap
    cmap = sns.color_palette(palette ='Blues_r', as_cmap = True)
    # Add a light gray color to represent NaN values (threshold not attained)
    cmap_with_nan = ListedColormap(["lightgray"] + list(cmap(np.linspace(0, 1, 256))))
    
    # Shift all epoch values by +1 for color mapping so that:
    # NaNs become 0 and map to light gray,
    # valid epochs start at 1 in the colormap scale
    plot_data = pivot.copy()
    plot_data += 1
    plot_data = plot_data.fillna(0)  # Replace NaNs with 0 to map to light gray
  
    # Create the heatmap figure
    plt.figure(figsize=(12, 6))
    ax = sns.heatmap(plot_data, 
                     cmap=cmap_with_nan, 
                     cbar=True, 
                     annot=pivot,        # Show the original epoch numbers as annotations
                     fmt=".1f",          # Format annotations as floats with 1 decimal
                     linewidths=0.5,
                     vmin=1,             # Start coloring valid data from index 1 in colormap
                     mask=None)

    # Customize colorbar label with the threshold percentage
    colorbar = ax.collections[0].colorbar
    colorbar.set_label(f"Epochs to reach {int(threshold*100)}% accuracy", fontsize=12)

    # Set plot titles and axis labels
    plt.title(f"Convergence Heatmap (Threshold = {int(threshold*100)}% Accuracy)", fontsize=14)
    plt.xlabel("Learning Rate", fontsize=12)
    plt.ylabel("Optimizer", fontsize=12)
    plt.xticks(rotation=45)   # Rotate x-axis labels for readability
    plt.tight_layout()        # Adjust layout to fit everything nicely
    plt.savefig(f"graphs/convergence.png")  # Save the figure as a PNG file
    plt.show()                # Display the heatmap plot

# Call the plotting function with the results and threshold
plot_convergence_heatmap(results, threshold=threshold)
```
